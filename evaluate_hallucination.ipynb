```python
from transformers import BertTokenizer, BertModel
from bert_score import BERTScorer
# Examples
reference = "The provision prohibits the court from reviewing the reason for the decision to deny a security clearance."
candidate = "The provision sayes that the court can't review the reason for the decision to deny a security clearance."
```

```python
# Load Bert Scorer
scorer = BERTScorer(model_type='bert-base-uncased')
```

```python
#Score
P, R, F1 = scorer.score([candidate], [reference])
print(f"BERTScore: \nPrecision: {P.mean():.4f} \nRecall: {R.mean():.4f} \nF1: {F1.mean():.4f}")
```
